Посчитаем, сколько книг вышло после 1 января 2000 года.

```python
query('''
SELECT COUNT(book_id) AS total_book_cnt
FROM books
WHERE publication_date >= '2000-01-01'
''')
```

Для каждой книги посчитаем количество обзоров и среднюю оценку.

```python
books_rating = query('''
WITH
book_ratings AS
(SELECT  book_id,
        ROUND(AVG(rating), 1) AS avg_rating
FROM ratings
GROUP BY book_id),

book_review_cnt AS
(SELECT  book_id,
        COUNT(review_id) AS review_cnt
FROM reviews
GROUP BY book_id)

SELECT  b.book_id,
        title,
        avg_rating,
        COALESCE(review_cnt, 0) AS review_cnt
FROM books AS b
LEFT OUTER JOIN book_ratings AS br ON b.book_id=br.book_id
LEFT OUTER JOIN book_review_cnt AS brc ON b.book_id=brc.book_id
ORDER BY avg_rating DESC
''')

with pd.option_context("display.max_rows", None, 'display.max_colwidth', None):
    display(books_rating)
```

Количество строк в запросе 1000, поэтому построим гистограммы для общего представления по среднему рейтингу и количеству обзоров.

```python
books_rating.avg_rating.hist(bins=10);
plt.axvline(x=books_rating.avg_rating.mean(), 
            color='red', 
            linestyle='--',
            label='Средний рейтинг книг: {:.2f}'.format(books_rating.avg_rating.mean()))
plt.title('Распределение среднего рейтинга книг')
plt.xlabel('Рейтинг')
plt.ylabel('Частотность')
plt.legend()
plt.show()
```

Средний рейтинг книг в базе равен 3,9. Большинство книг имеет рейтинг выше 3.

```python
books_rating.review_cnt.hist(bins=7);
plt.axvline(x=books_rating.review_cnt.mean(), 
            color='red', 
            linestyle='--',
            label='Среднее количество обзоров на книгу: {:.1f}'.format(books_rating.review_cnt.mean()))
plt.title('Распределение количества обзоров на книгу')
plt.xlabel('Количество обзоров на книгу')
plt.ylabel('Частотность')
plt.legend()
plt.show()
```

В среднем на книгу в базе приходится 2,8 обзора от читателей. Большинство книг получили от 2 до 4 обзоров.

Подсчитаем среднюю оценку и среднее количество обзоров по базе данных книг.

Определим издательство, которое выпустило наибольшее число книг толще 50 страниц — так мы исключим из анализа брошюры.

```python
query('''
SELECT  publisher,
        COUNT(b.book_id) AS total_cnt
FROM books AS b
LEFT OUTER JOIN publishers AS p ON b.publisher_id=p.publisher_id
WHERE num_pages > 50
GROUP BY publisher
ORDER BY total_cnt DESC
LIMIT 5
''')
```

Определим автора с самой высокой средней оценкой книг — учтем только книги с 50 и более оценками.

```python
query('''
WITH
avg_book_ratings AS 
(SELECT author_id,
        b.book_id,
        AVG(rating) AS avg_rating,
        COUNT(rating_id) AS rating_cnt
FROM ratings AS r
LEFT OUTER JOIN books AS b ON b.book_id=r.book_id
GROUP BY author_id, b.book_id
HAVING COUNT(rating_id) >= 50
ORDER BY avg_rating DESC),

avg_author_rating AS
(SELECT  author_id,
        AVG(avg_rating) AS avg_author_rating
FROM avg_book_ratings
GROUP BY author_id
ORDER BY avg_author_rating DESC)

SELECT  author,
        ROUND(avg_author_rating, 2) AS avg_rating
FROM authors as a
RIGHT OUTER JOIN avg_author_rating AS aar ON aar.author_id=a.author_id
ORDER BY avg_author_rating DESC
LIMIT 5
''')
```

Посчитаем среднее количество обзоров от пользователей, которые поставили больше 48 оценок.

```python
query('''
WITH
user_review_cnt AS
(SELECT username,
       COUNT(review_id) AS review_cnt
FROM reviews
WHERE username IN (SELECT username
                    FROM ratings
                    GROUP BY username
                    HAVING COUNT(rating_id) > 48)
GROUP BY username)

SELECT ROUND(AVG(review_cnt), 2) AS avg_review_cnt
FROM user_review_cnt

''')
```
